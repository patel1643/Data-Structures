## Solution for exercise 1

```python
# Python code and its output for "get and update" function

import timeit
import random
from matplotlib import pyplot as plt 


#for graph
yAxisForGet = []
xAxisForGet = []
yAxisForUpdate = []
xAxisForUpdate = []


for i in range(10000,1000001,20000):
    t_Get = timeit.Timer("x.get('random.randrange(i)')",
                     "from __main__ import random,x")
    
    t_Update = timeit.Timer("x.update({random.randrange(%d):None})"%i,
                     "from __main__ import random,x")
    
    x = {j:None for j in range(i)}
    
    dict_time_forGet = t_Get.timeit(number=1000)
    dict_time_forUpdate = t_Update.timeit(number=1000)
    
    xAxisForGet.append(i)
    yAxisForGet.append(dict_time_forGet)
    
    xAxisForUpdate.append(i)
    yAxisForUpdate.append(dict_time_forUpdate)

plt.figure(figsize = (10,5))
plt.plot(xAxisForGet, yAxisForGet,"r+")
plt.plot(xAxisForUpdate, yAxisForUpdate, "g^")
plt.title('Graph for get and update function runtime for Dictionaries')
plt.legend(['Get', 'Update'], loc = 0)
plt.figure
```

**Discussion of the Results**

This code is the answer for the first question in which we determine if the get and update methods in Dictionary data type are the same order of operation that is O(1). From the graph obtained we can clearly see that both the functions are indepentent of time and no matter how big the number gets it would not affect the runtimes of either of the functions/methods.
